```python
import pandas as pd
t_data=pd.read_csv("titanic_train.csv")
t_data
```

```python
t_data.drop("Cabin",inplace=True,axis=1)
t_data.drop("Name",inplace=True,axis=1)
t_data.drop("Ticket",inplace=True,axis=1)
t_data['Age'] = (t_data.groupby(['Survived','Sex','Pclass'])['Age'].transform(lambda x: x.fillna(x.mean())))
```

```python
t_data.shape
```

```python
sex=pd.get_dummies(t_data["Sex"],drop_first=True)
embark=pd.get_dummies(t_data["Embarked"],drop_first=True)
pclass=pd.get_dummies(t_data["Pclass"],drop_first=True)                                                        
```

```python
t_data.drop(["Pclass","Sex","Embarked"],inplace=True,axis=1)
```

```python
t_data
```

```python
t_data=pd.concat([t_data,sex,embark,pclass],axis=1)
t_data
```

```python
t_data.isnull().sum()
```

```python
data=t_data.drop("Survived",axis=1)
target=t_data["Survived"]
data
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(solver="saga")
clf.fit(data,target)
```
